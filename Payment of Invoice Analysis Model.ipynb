```python
df['Created_Day'] = df['Invoice_Date'].str.split('-' , n = 2 , expand = True)[0]
df['Created_Month'] = df['Invoice_Date'].str.split('-' , n = 2 , expand = True)[1]
df['Created_Year'] = df['Invoice_Date'].str.split('-' , n = 2 , expand = True)[2]
# df['Created_Hour'] = df['Created'].str.split('-' , n = 2 , expand = True)[2].str.split(' ', n=1,expand = True)[1].str.split(':' , n = 1 , expand = True)[0]
# df['Created_Minute'] = df['Created'].str.split('-' , n = 2 , expand = True)[2].str.split(' ', n=1,expand = True)[1].str.split(':' , n = 1 , expand = True)[1]
df
```

```python
df.drop(labels = ['Invoice_Date'] , axis = 1 , inplace = 
        True)
df
```

```python
df.drop(labels = ['Created'],axis=1,inplace= True)
df
```

```python
df['Due_Day'] = df['Due_Date'].str.split('-' , n = 2 , expand = True)[0]
df['Due_Month'] = df['Due_Date'].str.split('-' , n = 2 , expand = True)[1]
df['Due_Year'] = df['Due_Date'].str.split('-' , n = 2 , expand = True)[2]
df
```

```python
df.drop(labels = ['Due_Date'] , inplace = True,axis = 1)
df
```

```python
dict = {'Jan' : 1 , 'Feb' : 2 , 'Mar' : 3 , 'Apr' : 4 , 'May' : 5 , 'Jun' : 6 , 'Jul' : 7 , 'Aug' : 8 , 'Sep' : 9 , 'Oct' : 10 , 'Nov' : 11 , 'Dec' : 12}
df['Due_Month'] = df['Due_Month'].map(dict)
df['Created_Month'] = df['Created_Month'].map(dict)
df
```
