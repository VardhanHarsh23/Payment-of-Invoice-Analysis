```python
# Firstly I am starting with Created_Day , Created_Month , Created_Year from Invoice Date
```

```python
arr = df['Invoice_Date'].str.split('-' , n = 2 , expand = True)
arr
```

```python
# Now making different columns for them
df['Created_Day'] = arr[0]
df['Created_Month'] = arr[1]
df['Created_Year'] = arr[2]
```

```python
df
```

```python
# dropping the Invoice column
df.drop(labels = ['Invoice_Date'] , axis = 1 , inplace = True)
```

```python
df
```

```python
# Since Created and Invoice date are neraly same so I also remove the created Column
df.drop(labels = ['Created'] , axis = 1 , inplace = True)
```

```python
df
```

```python
# Same thing as done with Invoice date we would do with due_date
arr = df['Due_Date'].str.split('-' , n = 2 , expand = True)
arr
```

```python
# Now making different columns for them
df['Due_Day'] = arr[0]
df['Due_Month'] = arr[1]
df['Due_Year'] = arr[2]
```

```python
df
```

```python
df.drop(labels = ['Due_Date'] , axis = 1 , inplace = True)
df
```

```python
# Now Since my months are in string I wanted to convert them to numbers so I could easily do analysis on it
dict = {'Jan' : 1 , 'Feb' : 2 , 'Mar' : 3 , 'Apr' : 4 , 'May' : 5 , 'Jun' : 6 , 'Jul' : 7 , 'Aug' : 8 , 'Sep' : 9 , 'Oct' : 10 , 'Nov' : 11 , 'Dec' : 12}
df['Due_Month'] = df['Due_Month'].map(dict)
df['Created_Month'] = df['Created_Month'].map(dict)
df
```

```python
df['Description'].nunique()
```

```python
df['Vendor_Name'].nunique()
```

```python
# Since vendor name and description could be different in test dataset so I thought they are of no use in predicting a column. 
# Therefore I removed them
df.drop(labels = ['Vendor_Name' , 'Description'] , axis = 1 , inplace = True)
df
```

```python
df.dtypes
```

```python
# Since I would have to use all as numbers so there dtypes should not be object 
df = df.astype({'Created_Day': 'int64', 'Created_Year' : 'int64' , 'Due_Day' : 'int64' , 'Due_Year':'int64'})
df.dtypes
```

```python
# I also have a intution that that the no. of days (means difference in number of days between Due_date and Created_date)
df['diff_days'] = (df['Due_Day'] - df['Created_Day']) + (df['Due_Month'] - df['Created_Month']) * 30 + (df['Due_Year'] -df['Created_Year'])*365
df
```

```python
# Now as from data I know that Amount = Settled + Outstanding , So only any two columns are required to have all info about this data set regarding Bill payment
df.drop(labels = ['Amount'] , axis = 1 , inplace =True)
df
```

```python
outliers=[]
def detect_outliers(data):
    
    threshold=3
    mean = np.mean(data)
    std =np.std(data)
    
    
    for i in data:
        z_score= (i - mean)/std 
        if np.abs(z_score) > threshold:
            outliers.append(i)
    return outliers
```
